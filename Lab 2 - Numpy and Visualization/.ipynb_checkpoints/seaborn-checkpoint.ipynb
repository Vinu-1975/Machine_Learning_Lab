```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('iris')
data.sample(15)
```

```python
#data_sets available in sns
sns.get_dataset_names()
```

```python
sns.lineplot(x='sepal_length',y='sepal_width',data = data)
```

```python
sns.lineplot(x='sepal_length',y='sepal_width',data = data)
plt.title('Title using matplotlib func')
plt.xlim(4,8)
plt.show()
```

```python
sns.lineplot(x='sepal_length',y='sepal_width',data = data)
sns.set_style('whitegrid')
plt.show()
```

```python
sns.lineplot(x='sepal_length',y='sepal_width',data = data)
# sns.despine(left=True, right=True,offset=10) # by defalut remove top and right
plt.show()
```

```python
sns.lineplot(x='sepal_length',y='sepal_width',data = data)
plt.figure(figsize = (2,2))
sns.despine()
plt.show()
```
